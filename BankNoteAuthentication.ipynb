```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import random
import math
import statistics
import seaborn as sns
import matplotlib.pyplot as plt

#data from https://www.kaggle.com/ritesaluja/bank-note-authentication-uci-data#
#Data were extracted from images that were taken from genuine and forged banknote-like specimens#


df = pd.read_csv('data_banknote_authentication.txt', 
                                names=["variance", "skew", "curtosis", "entropy", "class"])
x = df.iloc[:,1]



sample_data = np.random.choice(a = x,size = 150)



point_estimators = []
for x in range(30):
    point_estimator = np.random.choice(a = sample_data, size = 4)
    point_estimators.append(point_estimator.mean())
    
    
sample_mean = statistics.mean(point_estimators)
sample_stdev = statistics.stdev(point_estimators)




sample_datapd = pd.DataFrame(sample_data, columns=["Variance of Wavelet Transformed Image"])
point_estimatorspd = pd.DataFrame(point_estimators, columns=["Point Estimators"])

point_estimatorspd['Point Estimators'].describe()
```

```python
f, axes = plt.subplots(2, 1, figsize=(10, 20))
sns.distplot( point_estimators , color="skyblue", ax=axes[0], label='Point Estimator')
sns.distplot( sample_data , color="olive", ax=axes[1], label="Sample Data")
```
